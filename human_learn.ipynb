добавляем еще одну переменную

```python
train.groupby('engineSize')['price'].agg(['count', 'mean', 'median'])
```

```python
def engine_group(x):
  if x <= 1.5:
    return '<=1.5'
  if x <= 2:
    return '<=2'
  return '>2'
```

```python
train['engine_group'] = train['engineSize'].apply(lambda x: engine_group(x))
```
